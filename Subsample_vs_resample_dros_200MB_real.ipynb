```python
import math
def jc(b, do_correction, a):
    deps = 0.0001
    if abs(b) < deps:
        return 0
    loc = 1 - (4.0 * b / 3)

    if (loc <= 0):
        return 5.0 # max dist
    if do_correction:
        if a < deps:
            a = deps
        ret = a * (0.75 * (loc**(-1.0/a) - 1))
    else:
        ret = -0.75 * (math.log(loc))
    return ret
```

JC correction

```python
import math
def jc(b, do_correction, a):
    deps = 0.0001
    if abs(b) < deps:
        return 0
    loc = 1 - (4.0 * b / 3)

    if (loc <= 0):
        return 5.0 # max dist
    if do_correction:
        if a < deps:
            a = deps
        ret = a * (0.75 * (loc**(-1.0/a) - 1))
    else:
        ret = -0.75 * (math.log(loc))
    return ret
```
